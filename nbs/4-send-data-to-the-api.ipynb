```python
import requests
```

```python
path = "/api/events"
base_url = "http://localhost:8002"
create_endpoint = f"{base_url}{path}"

response = requests.post(create_endpoint, json={"page":"/test"})
print('Response ok:', response.ok)

if response.ok :
    data = response.json()
    print(data, type(data), data.get('items'))
else:
    print(response.text)
    
```

```python
detail_path = "/api/events/1"
detail_endpoint = f"{base_url}{detail_path}"

response = requests.put(detail_endpoint, json={"description":"Hello world"})
print(response.headers)
print('ok', response.ok)

if response.ok :
    data = response.json()
    print(data, type(data), data.get('items'))
else:
    print(response.text)
    
```

```python
import random
event_count = 10_000

pages = ['/about', '/contact', '/pages', '/pricing']

for i in range(event_count):
    page = random.choice(pages)
    response = requests.post(create_endpoint, json={"page":page})
    if response.ok:
        data = response.json()
        print(data, type(data), data.get('items'))
    else:
        print(response.ok)
```

```python
from faker import Faker
import random

events = 10_000
pages = [
        "/", "/about", "/pricing", "/contact", 
        "/blog", "/products", "/login", "/signup",
        "/dashboard", "/settings"
    ]
fake = Faker()
session_ids = [fake.uuid4() for _ in range(20)]
path = "/api/events/"
base_url = "http://localhost:8002" # 127.0.0.1
create_endpoint = f"{base_url}{path}"
referrers = [
    "https://google.com",
    "https://facebook.com",
    "https://twitter.com",
    "https://linkedin.com",
    "",
    "https://github.com"
]

for i in range(events):
    page = random.choice(pages)
    user_agent = random.choice([
            fake.chrome,
            fake.firefox,
            fake.safari,
            fake.opera,
            fake.user_agent
        ])()
    payload = dict(
        page=random.choice(pages),
        user_agent=user_agent,
        ip_address=fake.ipv4(),
        referrer=random.choice(referrers),
        session_id=random.choice(session_ids),
        duration=random.randint(1, 300)
    )
    response = requests.post(create_endpoint, json=payload)
    
    if response.ok:
        data = response.json()
        print(data)
```
